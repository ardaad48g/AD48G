Setting seed to 90, calling necessary libraries and generating 'random_dataset' whic includes: Letters, numbers and some special characters.

```r
set.seed(90)
library(data.table)
library(tidyverse)
library(htm2txt)
library(pdftools)
library(textreadr)
library(magrittr)


generate_random_strings <- function(n, length) {
  chars <- c(letters, LETTERS, 0:9, "!@#$%^&*()")
  result <- sapply(1:n, function(i) {
    paste(sample(chars, length, replace = TRUE), collapse = "")
  })
  return(result)
}

random_dataset <- generate_random_strings(100, 20)
```

Using str_extract with '+ Quantifier' (It extracts if there is 'a' or more 'a's in variable)

```r
random_dataset %>% str_extract("a+")
```

Using str_extract_all with '* Quantifier' (It extracts all 1's and '2's if there is any before '1')

```r
str_extract_all(random_dataset, "2*1") 
```

Using str_extract_all with '? Quantifier' (It is same like * but it does not take all 2's, it takes if there is zero or one of first variable)

```r
str_extract_all(random_dataset, "2?1") 
```

Using str_extract_all with '. Qualifier' (With '.' , we extracted all variable that starts with g and ends with a)

```r
random_dataset %>% unlist %>% str_extract_all("g.*a")
```

Using str_replace with '\w Quanlifier' and '^ Anchor' (We replaced zero or many word characters with '1' from the begging of the variable(with '^') )

```r
str_replace(random_dataset, "^\\w*", "1") 
```

Using str_detect with '\d Qualifier' with '$' (We detected if variable is finishing with one or more numbers)

```r
str_detect(random_dataset, "\\d+$")
```

Using str_replace_all with '\s Qualifier' (With \s Qualifier we can remove spaces but, for our dataset we see that there is no space to replace) 

```r
str_replace_all(random_dataset, "\\s+", "AAAA")
```

Using str_replace_all with '[]' (With [] we only keep letters and replaced non letters with "") 

```r
str_replace_all(random_dataset, "[^a-zA-Z]+", "")
```

Using str_detect with 'I (or)' (We detected if variables in dataset contains 'a' or '4')

```r
str_detect(random_dataset, "a|4")
```

Excluded character set defined inside [^] (We replaced any char that is not included btw 0-9, so we replaced all non number characters with "")

```r
str_replace_all(random_dataset, "[^0-9]", "")
```

(?<=) positive lookbehind (pattern following by another pattern)  (We extracted any numbers that are next to 'f')

```r
str_extract(random_dataset, "(?<=f)\\d+")
```

(?=) positive look ahead (pattern followed by another pattern) (It extracts the left part from the last digit of the variable)

```r
str_extract(random_dataset, "\\w+(?=\\d+)")
```
